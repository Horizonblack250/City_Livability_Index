```python
import pandas as pd
import os

RAW = r"C:\Users\alokhande\Desktop\Project_Daedalus\data\raw"

datasets = {
    "AQI":              os.path.join(RAW, "global air pollution dataset.csv"),
    "Cost_of_Living_v2": os.path.join(RAW, "cost-of-living_v2.csv"),
    "Pop_Density_1":    os.path.join(RAW, "CCCdata1.csv"),
    "Pop_Density_2":    os.path.join(RAW, "CCCdata2.csv"),
    "Crime_Index":      os.path.join(RAW, "crime-index-2023.csv"),
}

for name, path in datasets.items():
    print(f"\n{'='*60}")
    print(f"  {name}")
    print(f"{'='*60}")
    try:
        df = pd.read_csv(path, encoding='utf-8')
    except UnicodeDecodeError:
        df = pd.read_csv(path, encoding='latin-1')

    print(f"Shape        : {df.shape[0]} rows x {df.shape[1]} cols")
    print(f"\nColumns ({df.shape[1]}):")
    for col in df.columns:
        print(f"  - {col}  [{df[col].dtype}]  | nulls: {df[col].isnull().sum()}")
    print(f"\nFirst 3 rows:")
    print(df.head(3).to_string())
    print(f"\nBasic stats (numeric only):")
    print(df.describe().to_string())
```
